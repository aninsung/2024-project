```python
import string


src_str = string.ascii_uppercase
dst_str= src_str[3:] +src_str[:3]

def ciper(a):
    idx= src_str.index(a)
    return dst_str[idx]
src=input("문장을 입력하시오:")
print("암호화된 문장:",end='')


for ch in src:
    if ch in src_str:
        print(ciper(ch), end='')
    else:
        print(ch,end='')


print()
```

```python
import re

while True:
    password = input("패스워드를 입력하시오: ")
    if len(password) < 8 or not re.search("[a-z]", password) or \
       not re.search("[A-Z]", password) or \
       not re.search("[0-9]", password) or not re.search("[_@$!]", password):
        print("유효하지 않은 패스워드!")
    else:
        print("유효한 패스워드")
        break
```
